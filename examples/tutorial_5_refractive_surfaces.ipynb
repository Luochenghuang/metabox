Modeling refractive surfaces.

Here, we optimize two even asphere surfaces in Zemax.  
We use BK7 as the refractive medium between them.
![zemax](img/zemax_1.png)

```python
from metabox import assembly

incidence = assembly.Incidence(
    wavelength=[700e-9],
    theta=[0],
    phi=[0],
)

asph1 = assembly.RefractiveEvenAsphere(
    diameter=0.1e-3,
    refractive_index=1.5131,
    thickness=0.1e-3,
    periodicity=0.25e-6,
    init_coeff=[6.124, 97.618],
)

asph2 = assembly.RefractiveEvenAsphere(
    diameter=0.2e-3,
    refractive_index=1.0,
    thickness=0.1e-3,
    periodicity=0.25e-6,
    init_coeff=[0.731, -189.048],
)

# Create a lens assembly.
lens_assembly = assembly.LensAssembly(
        surfaces=[asph1, asph2],
        incidence=incidence,   # Define the incidence.
        figure_of_merit=assembly.FigureOfMerit.LOG_STREHL_RATIO, # Define the figure of merit.
)

asph1.show_sag()
asph2.show_sag()
lens_assembly.show_psf()
```

As you can see, `metabox` is capable of modeling simple refractive systems.
What about more complex systems?

Here we optimize a four surface refractive assembly using Zemax Again, we are using four even asphere surfaces with BK7 as the refractive medium.
![zemax](img/zemax_2.png)

```python
from metabox import assembly

incidence = assembly.Incidence(
    wavelength=[700e-9],
    theta=[0, 7.5, 15],
    phi=[0],
)

asph1 = assembly.RefractiveEvenAsphere(
    diameter=0.1e-3,
    refractive_index=1.5131,
    thickness=0.1e-3,
    periodicity=0.25e-6,
    init_coeff=[2.916, 11.174],
)

asph2 = assembly.RefractiveEvenAsphere(
    diameter=0.2e-3,
    refractive_index=1.0,
    thickness=0.1e-3,
    periodicity=0.25e-6,
    init_coeff=[-0.105, -10.073],
)

asph3 = assembly.RefractiveEvenAsphere(
    diameter=0.1e-3,
    refractive_index=1.5131,
    thickness=0.1e-3,
    periodicity=0.25e-6,
    init_coeff=[-0.078, -65.907],
)

asph4 = assembly.RefractiveEvenAsphere(
    diameter=0.2e-3,
    refractive_index=1.0,
    thickness=0.1e-3,
    periodicity=0.25e-6,
    init_coeff=[0.323, 134.355],
)

# Create a lens assembly.
lens_assembly = assembly.LensAssembly(
        surfaces=[asph1, asph2, asph3, asph4],
        incidence=incidence,   # Define the incidence.
        figure_of_merit=assembly.FigureOfMerit.LOG_STREHL_RATIO, # Define the figure of merit.
)

lens_assembly.show_psf()    
```

As you can see, `metabox` is capable of modeling after complex refractive optical systems with a reasonable accuracy.
