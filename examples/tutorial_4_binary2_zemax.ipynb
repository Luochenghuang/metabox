Importing Binary 2 surfaces from Zemax.

Here, we optimize a doublet system with  
(1) a 50µm diameter Binary 2 surface, 0.1mm thick.  
(2) a 100µm diameter Binary 2 surface, 0.2mm thick.  
The angles of incidence are 0, 5, and 10°.  
![zemax](./img/zemax_0.png)

```python
from metabox import assembly

incidence = assembly.Incidence(
    wavelength=[700e-9],
    theta=[0, 5, 10],
    phi=[0],
)

# Define the binary surface with the coefficients.
bin1 = assembly.Binary2(
    diameter=0.1e-3,
    refractive_index=1.0,
    thickness=0.1e-3,
    periodicity=0.25e-6,
    init_coeff=[6.763, 2.1, -0.032, 0.055, -7.122e-3],
)

bin2 = assembly.Binary2(
    diameter=0.2e-3,
    refractive_index=1.0,
    thickness=0.2e-3,
    periodicity=0.25e-6,
    init_coeff=[-249.003, -10.79, -0.16, 0.24, 0.12],
)

# Create a lens assembly.
lens_assembly = assembly.LensAssembly(
        surfaces=[bin1, bin2], # Define the array of surfaces. Here only one.
        incidence=incidence,   # Define the incidence.
)

lens_assembly.show_psf()
```
